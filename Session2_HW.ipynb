```r
w0 <- 100 # Initial wealth
w0 <- 100

T <- 10000
time_periods <- T+1

w <- rep(0, T+1)
w[1] <- w0

X <- rep(0, T)
i <- 1

p <- runif(1)

while(T > 0){
    x <- runif(1)
    #print(sprintf("You started with %s euros.", w[i]))

    if(x > p){
        w[i+1] <- w[i] - 1
        X[i] <- 0
        #print("you lose")
    } else {
        w[i+1] <- w[i] + 1
        X[i] <- 1
        #print("you win")
    }
    # Comment...
    #print(sprintf("No youve got %s euros.", w[i+1]))
    
    T <- T - 1
    i <- i + 1
}
plot(1:time_periods, w, type='l')
```
